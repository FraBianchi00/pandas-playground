### Osservazioni

- Il dataset ha 891 righe e 12 colonne
- Valori mancanti:
  - `Cabin`: 687 su 891 (~77%) — colonna largamente inutilizzabile nella forma attuale
  - `Age`: 177 su 891 (~20%) — significativo, da trattare (non semplicemente eliminare)
  - `Embarked`: 2 su 891 (~0,2%) — trascurabile, si possono eliminare le 2 righe
- `Age` è `float64` e non `int64` a causa dei NaN: in pandas i valori mancanti
  forzano il tipo float
