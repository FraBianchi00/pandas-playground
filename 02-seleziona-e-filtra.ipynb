## Sezione 3: Selezionare righe (.loc)

```python
titanic.loc[5:10]
```

## Sezione 4: Maschere booleane

```python
titanic["Survived"] > 0
```

```python
titanic[titanic["Survived"] == True]
```

## Sezione 5: Combinare condizioni

```python
titanic[(titanic["Sex"] == "female") & (titanic["Pclass"] == 1)]
```

```python
titanic[(titanic["Age"] < 18) & (titanic["Survived"] == True)]
```

```python
len(titanic[(titanic["Survived"] == 1) & (titanic["Age"] < 18)])/len(titanic[titanic["Age"] < 18])
```

## Osservazioni

Il tasso generale di sopravvivenza si aggira attorno al 38%. Confrontandolo con il tasso di bambini sopravvissuti (pari al 53%), il che implica una maggior probabilità di sopravvivenza per le persone sotto i 18 anni.
