## Sezione 1: Warm up

```python
import pandas as pd
```

```python
titanic = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
titanic["Embarked"].value_counts()
```

```python
titanic["Sex"].value_counts(normalize=True)
```

```python
titanic["Survived"].value_counts()
```

## Sezione 2: Groupby

```python
titanic.groupby("Pclass")["Survived"].mean()
```

```python
titanic.groupby("Sex")["Survived"].mean()
```

```python
titanic.groupby(["Pclass", "Sex"])["Survived"].mean()
```

```python
titanic.groupby("Pclass")["Age"].mean()
```

## Sezione 3: .agg() 

```python
titanic.groupby("Pclass")["Age"].agg(["mean", "median", "std"])
```

```python
titanic.groupby(["Pclass", "Sex"]).agg(
    eta_media=("Age", "mean"), 
    fare_medio=("Fare", "mean"), 
    sopravvissuti=("Survived", "sum"),
    totale=("Survived", "size"))
```

```python
titanic.groupby(["Pclass", "Sex"])["Age"].mean()
```
